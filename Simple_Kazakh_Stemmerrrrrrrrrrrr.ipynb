```python
texts = text[0]

#remove punctuation

def rem_punc(t):
    # define punctuation
    punctuations = '''!()[]{};:'"\,<>./?@#$%^&*_~abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890'''

    #remove punctuation from the string
    for c in punctuations:
       t = t.replace(c,' ')


    t = t.replace('\n',' ')

    #display the unpunctuated string
    return t
texts = rem_punc(texts)
```

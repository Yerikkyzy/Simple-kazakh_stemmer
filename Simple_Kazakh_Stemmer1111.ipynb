```python
def word_encode(word): # encode given word to the form of C and V(consonant and vowel)
    syl = ''
    
    for i in word:
        if i in vowel:
            syl+='V'
        else:
            syl+='C'

    return syl

def get_syllable_sequence(word):  # Main function: dividing encoded words to syllables
    syl_seq = word_encode(word)
    
    
    while 'VCC' in syl_seq:
        syl_seq = syl_seq.replace("VCC","VC-C")
    while "CVCV" in syl_seq:
        syl_seq = syl_seq.replace('CVCV', 'CV-CV')
    while 'CVVCVC' in syl_seq:
        syl_seq = syl_seq.replace('CVVCVC', 'CV-V-CVC')
    while 'CVVVC' in syl_seq:
        syl_seq = syl_seq.replace('CVVVC', 'CV-V-VC')
    while 'VCV' in syl_seq:
        syl_seq = syl_seq.replace('VCV', 'V-CV')
    while 'CVVCCVC' in syl_seq:
        syl_seq = syl_seq.replace('CVVCCVC','CV-VC-CVC')
        

    return syl_seq

```

```python
def text_syllable(text): # dividing whole encoded text into syllables
    sylls = []
    prepared = []
    
    for word in text:
        syll = get_syllable_sequence(word)
        sylls.append(syll)
        
    for syl in  sylls:
        prepared.append(syl.split('-'))
    
    return prepared
        
```

```python
def Stemmatizer_word(word): # decoding every splitted syllables into words
    word_index = prepared_text.index(word)
    t = []
    create_root = ''

    for syl in syllabled_text[word_index]: 
        p = word[:len(syl)]
        word = word.replace(word[:len(syl)], '')
        t.append(p)
    
    for i in t:
        create_root+=i
        if create_root in corpus:
            break
            
    return create_root
```

```python
def get_roots(text): # get roots of whole text
    complete_text = []
    
    for word in text:
        root_word = Stemmatizer_word(word)
        complete_text.append(root_word)
        
    return complete_text
        
```
